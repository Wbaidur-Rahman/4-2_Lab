**Write a Python program to display values of y when 'x' is an independent variable having values [-10, 10, step_size = 0.1] and:**

```python
import numpy as np
import matplotlib.pyplot as plt
```

# y = x^2

```python
x = np.arange(-10, 10, 0.1)
y = x**2

plt.plot(x, y)
plt.xlabel('X-Value')
plt.ylabel('Y-Value')
```

**y = 1/ (1 + e^(-x));**

```python
y = 1/ (1 + np.exp(-x))
plt.plot(x, y)
plt.xlabel('X-Value')
plt.ylabel('Y-Value')
```

**y = (e^x - e^(-x)) / (e^x - e^(-x))**


```python
y = (np.exp(x) - np.exp(-x)) / (np.exp(x) - np.exp(-x))

plt.plot(x, y)
plt.xlabel('X-Value')
plt.xlabel('Y-Value')
```

**y = g(f(x)) where u = f(x) = wx + b and y = g(u) = 1/ (1 + e^(-u))**

```python
w, b = 8, 5
u = w*x + b
y = 1/(1 + np.exp(-u))


plt.plot(x, y)
plt.xlabel('X-Value')
plt.xlabel('Y-Value')
```

**y = g(f(x)) where f(x) = wx + b, u = f(x), and g(u) = (e^u - e^(-u)) / (e^u + e^(-u))**

```python
w, b = 7, 9
u = w*x + b
y = (np.exp(u) - np.exp(-u)) / (np.exp(u) - np.exp(-u))

plt.plot(x, y)
plt.xlabel('X-Value')
plt.xlabel('Y-Value')
```

**y1 = g1(f1(x)) where f1(x) = w1x + b1, u1 = f1(x), and g1(u1) = 1/ (1 + e^(-u1))**

```python
w, b = 3, 5
u = w*x + b
y1 = 1/ (1 + np.exp(-u))

plt.plot(x, y1)
plt.xlabel('X-Value')
plt.ylabel('Y-Value')
```

**y2 = g2(f2(x)) where f2(x) = w2x + b2, u2 = f2(x), and g2(u2) = 1/ (1 + e^(-u2))**

```python
w, b = 7, 11
u = w*x + b
y2 = 1/ (1 + np.exp(-u))

plt.plot(x, y2)
plt.xlabel('X-Value')
plt.ylabel('Y-Value')
```

**y = g3(f3(v)) where f3(v) = w3y1 + w4y2 + b, w = f3(v), and g3(w) = 1/ (1 + e^(-w))**

```python
b = 11
w3, w4 = 7, 10
w = w3*y1 + w4*y2 + b
y = 1/ (1 + np.exp(-w))

plt.plot(x, y)
plt.xlabel('X-Value')
plt.ylabel('Y-Value')
```
