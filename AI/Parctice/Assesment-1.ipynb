```python
import matplotlib.pyplot as plt
import math

stepSize = 0.001
limit = [-10, 10]
x = []
value = limit[0]
a = 8
b = 10

while value <= limit[1]:
    x.append(value)
    value += stepSize

# plt.plot(x)
print(x)
```

**y = ax + b**

```python
y = []
for val in x:
    y.append(a*val+b)

plt.ylabel('Y-Value')
plt.xlabel('X-Value')
plt.plot(x, y)
```

**y = a*sin(x) + b**

```python
y = []
for val in x:
    y.append(a*math.sin(val)+b)

plt.ylabel('Y-Value')
plt.xlabel('X-Value')
plt.plot(x, y)
```

**y = a*e^x + b**

```python
y = []
for val in x:
    y.append(a*math.exp(val) + b)

plt.ylabel('Y-Value')
plt.xlabel('X-Value')
plt.plot(x, y)
```

**y = a*x^2 + b**

```python
y = []
for val in x:
    y.append(a*val*val+b)

plt.ylabel('Y-Value')
plt.xlabel('X-Value')
plt.plot(x, y)
```
